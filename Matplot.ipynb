```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
hor=np.arange(0,100)
ver = np.arange(0,100)
```

```python
fig, ax = plt.subplots()
ax.barh(hor, ver)
```

```python
fig, ((ax1,ax2),(ax3,ax4)) =  plt.subplots(nrwos=2,ncols=2,figsize=(20,10))
ax1.plot(hor,hor**2, label="hello")
ax2.scatter(ver,ver**2, label="bye")
ax3.bar(hor,hor**3, color='brown')
ax4.plot(ver,ver**3, color='red')
fig.legend();
```
