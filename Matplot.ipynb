```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
hor=np.arange(0,100)
ver = np.arange(0,100)
```

```python
fig, ax = plt.subplots()
ax.barh(hor, ver)
```

```python
fig, ((ax1,ax2),(ax3,ax4)) =  plt.subplots(nrows=2,ncols=2,figsize=(20,10))
ax1.plot(hor,hor**2, label="hello")
ax2.scatter(ver,ver**2, label="bye")
ax3.bar(hor,hor**3, color='brown')
ax4.plot(ver,ver**3, color='red')
fig.legend();
```

```python
df = pd.read_csv('heart-disease.csv')
df.head()
```

```python
df['chol'].plot.hist();
#df['chol'].plot(kind='hist')
```

```python
df.plot.hist(subplots=True, figsize=(10,30));
```

```python
over_50 = df[df['age']>50]
over_50.head()
```

```python
## Pyplot method
over_50.plot(kind='scatter', x='age',y='chol',c='target',figsize=(10,6));
```

```python
## OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age',y='chol',c='target', ax=ax)
ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

## Plot the data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

## Customize the plot
ax.set(title='Heart Disease & Cholesterol Levels', xlabel='Age', ylabel='Cholesterol')

## Add legend
ax.legend(*scatter.legend_elements(),title='Target')

## Add horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
fig,(ax1,ax2) = plt.subplots(nrows=2, ncols=1,figsize=(10,10))
scatter = ax1.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
scatter = ax2.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])

ax1.set(title='Heart Disease & Cholesterol Levels', ylabel='Cholesterole')
ax2.set(title='Heart Disease & Thalach Levels', xlabel='Age', ylabel='Thalach')

fig.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(y=over_50['chol'].mean())
ax2.axhline(y=over_50['thalach'].mean(),color='red')

fig.suptitle('Main data', fontsize=16, fontweight='bold');
```
